```python
import torch, pickle
import matplotlib.pyplot as plt
from torch import cos

pi = torch.pi
with open('/home/dgotzens/scripts/2.2_antenna-characteristics/measured/a08_out.pkl', 'rb') as f:
    angle, level = pickle.load(f)

theta_r = torch.tensor(angle)-pi/2
plt.plot(180/pi * theta_r, level, label='original')
plt.plot(180/pi * theta_r, 20*cos(theta_r), label=r'$20cos\theta_r$')
plt.plot(180/pi * theta_r, 20*(cos(theta_r)**2), label=r'$20(cos\theta_r)²$')
plt.grid()
plt.legend()
plt.show()
```
